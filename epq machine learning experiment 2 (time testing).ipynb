```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn import svm
from sklearn import neighbors
from sklearn import tree
import random
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc  ###计算roc和auc
from sklearn import model_selection
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
obj = load_iris()
print(iris)
```

```python
trainning_num=100
testing_num=50
X_train=[]
y_train=[]
X_test=[]
y_test=[]
total_train=[]
total_test=[]
order=[]
for a,b in zip(iris.data,iris.target):
    total_train.append(a)
    total_test.append(b)
# total_train,total_test = total_train[total_test != 2],total_test[total_test != 2]
for i in range(len(obj.data)):
    order.append(i)
random.shuffle(order)
for i in range(trainning_num):
    X_train.append(total_train[order[i]])
    y_train.append(total_test[order[i]])
for i in range(testing_num):
    i = trainning_num+i
    X_test.append(total_train[order[i]])
    y_test.append(total_test[order[i]])

X_train=np.array(X_train)
y_train=np.array(y_train)
X_test=np.array(X_test)
y_test=np.array(y_test)
for k in [X_train,y_train,X_test,y_test]:
#     k = np.array(k)
    print(k)
```

```python
total_test=[]
for a,b in zip(iris.data,iris.target):
    total_train.append(a)
    total_test.append(b)
print(type(total_test))
```

```python
classifiers = [
    KNeighborsClassifier(),
    SVC(), 
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    MLPClassifier(max_iter=1000),
    GaussianNB()]
names = ["Nearest Neighbors", "SVM",  
         "Decision Tree", "Random Forest", "Neural Net",
         "Naive Bayes"]
```

```python


def auc_calculate(clf):
    if hasattr(clf, "decision_function"):
        y_score= clf.fit(X_train, y_train).decision_function(X_test)
    else:
        y_score = clf.fit(X_train, y_train).predict_proba(X_test)[:, 1]
    fpr,tpr,threshold = roc_curve(y_test, y_score) ###计算真正率和假正率
    roc_auc = auc(fpr,tpr) ###计算auc的值
    return [roc_auc,fpr,tpr]



def roc_plt(data): 
    roc_auc=data[0]
    fpr=data[1]
    tpr=data[2]
    plt.figure()
    lw = 2
    plt.figure(figsize=(10,10))
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc) ###假正率为横坐标，真正率为纵坐标做曲线
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    
```

```python
def accuracy_calculate(clf):
    clf = clf.fit(X_train, y_train)
    num = 0
    for a,b in zip(clf.predict(X_test),y_test):
        if a==b:
            num=num+1
                
    return(num/testing_num)
```

```python
import time
for name,clf in zip(names, classifiers):
    print(name)
    times = 100
    total=0
    start =time.clock()
    for i in range(times): 
        total = total+accuracy_calculate(clf)
    total = total/times
    end = time.clock()
    print('accuracy: %f' %(total))
    print('Running time: %s Seconds'%(end-start))       
```
